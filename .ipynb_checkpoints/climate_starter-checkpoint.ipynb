# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
 
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
from flask import Flask,jsonify
```

```python
app = Flask(__name__)
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite",echo=False)
#engine = create_engine("sqlite:///../Resources/emoji.sqlite", echo=False)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
records = session.query(Measurement)
for rec in records: 
    print (rec.date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results


# Calculate the date 1 year ago from today
last_yr = dt.datetime.now() - dt.timedelta(days=1*365)
today= dt.datetime.now()


# Perform a query to retrieve the data and precipitation scores
percp_data = engine.execute("SELECT date, prcp FROM measurement WHERE date BETWEEN '2016-08-01' AND '2017-08-01'").fetchall()

# Save the query results as a Pandas DataFrame and set the index to the date column
per_df = pd.DataFrame(percp_data)
per_df = per_df.rename(columns={0:'date',1:'percipitation'})
per_df.set_index('date')

# Sort the dataframe by date
per_df.sort_values('date')
per_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

# Rotate the xticks for the dates
plt.figure(num=1)
plt.title("Percipitation from August 2016 to August 2017")
plt.xlabel("Date")
plt.ylabel("Percipitation")
plt.grid()

xaxis = per_df["date"]
yaxis = per_df["percipitation"]

#xaxis = df["percipitation"]
#yaxis = df["date"]

plt.scatter(xaxis,yaxis,edgecolor='blue', facecolor = 'green', marker ='^')
plt.savefig("./Images_SurfUp/PrcpAnalysis.png")
plt.show()
```

```python
#STATION ANALYSIS

# How many stations are available in this dataset?
numStations = session.query(Station).count()

# What are the most active stations?
# List the stations and the counts in descending order.
active = session.query(Measurement.station,func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
   
highest = session.query(Measurement.station,func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).\
    having(func.max(func.count(Measurement.station)))

highest

print(f'There are {numStations} stations')

df_one = pd.DataFrame(active, columns=['StationName', 'Activity'])

print(f'\n The list of stations are as follows \n\n {df_one}')

highest = df_one.max()

print(f'\n The station with the highest number of observation is: \n {highest}')
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station

calTemp = session.query(Measurement.station,func.count(Measurement.station),func.max(Measurement.tobs),func.min(Measurement.tobs),func.avg(Measurement.tobs)).\
     group_by(Measurement.station).\
     order_by(func.count(Measurement.station).desc()).all()
      
#calTemp

df_temp = pd.DataFrame(calTemp, columns=['StationName', 'Activity','Max Temp', 'Min Temp','AvgTemp'])
df_temp
```

```python
# Choose the station with the highest number of temperature observations.
hot_station = session.query(Measurement.date,func.max(Measurement.tobs),func.min(Measurement.tobs)).all()
hot_station
    
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

year_ago = dt.date.today() - dt.timedelta(days=1*365)
today = dt.date.today()
    
tempQ2 = session.query(Measurement.tobs, func.count(Measurement.tobs)).\
filter(Measurement.date >=year_ago).\
filter(Measurement.date<=today).\
group_by(Measurement.tobs).all()    

df_tempQ2 = pd.DataFrame(tempQ2, columns=['Temp','Frequency'])
df_tempQ2
```

```python
plt.figure(num=2)
plt.title('Temperature Observations Frequency for time period')
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.grid()

xaxis = df_tempQ2["Temp"]
yaxis = df_tempQ2["Frequency"]

#xaxis = df["percipitation"]
#yaxis = df["date"]

plt.bar(xaxis,yaxis,edgecolor='white', facecolor = 'pink')
plt.savefig("./Images_SurfUp/StationAnalysis.png")
plt.show()
```
